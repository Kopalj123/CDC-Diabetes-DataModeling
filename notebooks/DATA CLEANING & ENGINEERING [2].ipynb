#### Define Functions for Data Cleaning

```python
# Create Function for sorting out Yes/No Features
def cleanFeature(num):
    if num == 1:
        return 1
    elif num == 2:
        return 0
    else:
        return np.nan
```

#### Define Functions for Data Engineering

```python
# Create Function for sorting out SLEEPTIME into SLEEPTIME_GROUP
def getSleepTimeGroup(time):
    if time <= 6:
        return 'LessSleep'
    elif time >= 7 and time <= 9:
        return 'NormalSleep'
    elif time >= 10:
        return 'OverSleep'
    else:
        return np.nan
```

```python
# Create Function for sorting out SEX into SEX_GROUP
def getSexGroup(num):
    if num == 1:
        return 'Male'
    elif num == 2:
        return 'Female'
    else:
        return np.nan
```

```python
# Create Function for converting WEIGHT into WEIGHT_KILOGRAM
def getWeightInKilogram(num):
    if np.isnan(num):
        return np.nan
        
    num = str(int(num))
    
    if num[0] == '9' and num != '9999':
        return float(num[1:])
    elif num != '9999' and num != '7777':
        return round(0.453592 * float(num), 3)
    else:
        return np.nan
```

```python
# Create Function for converting HEIGHT into HEIGHT_METER
def getHeightInMeter(num):
    if np.isnan(num):
        return np.nan
    
    num = str(int(num))

    if num[0] == '9' and num != '9999':
        return float(num[1:]) / 100 # centimeter to meter
    elif num != '9999' and num != '7777':
        return round((float(num[0])*12 + float(num[1:]))*0.0254, 3)
    else:
        return np.nan
```

```python
# Create Function for sorting out BMI into BMI_GROUP
def getBMIGroup(bmi):
    if bmi < 1850:
        return 'UnderWeight'
    elif bmi >= 1850 and bmi < 2500:
        return 'NormalWeight'
    elif bmi >= 2500 and bmi < 3000:
        return 'OverWeight'
    elif bmi >= 3000 and bmi < 9999:
        return 'Obese'    
    else:
        return np.nan
```

```python
# Create Function for sorting out RACE into RACE_GROUP
def getRaceGroup(num):
    if num == 1:
        return 'White'
    elif num == 2:
        return 'Black'
    elif num == 3:
        return 'Asian'
    elif num == 4:
        return 'AmericanIndian'
    elif num == 5:
        return 'Hispanic' 
    elif num == 6:
        return 'Other'
    else:
        return np.nan
```

```python
# Create Function for sorting out AGE into AGE_GROUP
def getAgeGroup(age):
    if age < 4:                 # 1: 18-24, 2: 25-29, 3: 30-34
        return 'YoungAdult'         
    elif age >= 4 and age < 7:  # 4: 35-30, 5: 40-44, 6: 45-49   
        return 'Adult'              
    elif age >= 7 and age < 10: # 7: 50-54, 8: 55-59, 9: 60-64
        return 'OldAdult'  
    elif age >= 10:             # 10: 65-69 11: 70-74, 12: 75-79, 13: 80+
        return 'Elderly'   
    else:
        return np.nan
```

#### Clean Existing Columns

```python
# Clean existing features
for data in [df_clean]:
    
    # Clean EXERCISE Feature
    data['EXERCISE'] = data['EXERCISE'].apply(lambda x: cleanFeature(x))
    
    # Clean HEARTATTACK Feature
    data['HEARTATTACK'] = data['HEARTATTACK'].apply(lambda x: cleanFeature(x))
    
    # Clean CORONARYHEARTDISEASE Feature
    data['CORONARYHEARTDISEASE'] = data['CORONARYHEARTDISEASE'].apply(lambda x: cleanFeature(x))
    
    # Clean STROKE Feature
    data['STROKE'] = data['STROKE'].apply(lambda x: cleanFeature(x))
    
    # Clean ASTHMA Feature
    data['ASTHMA'] = data['ASTHMA'].apply(lambda x: cleanFeature(x))
    
    # Clean ASTHMANOW Feature
    data['ASTHMANOW'] = data['ASTHMANOW'].apply(lambda x: cleanFeature(x))
    
    # Clean SKINCANCER Feature
    data['SKINCANCER'] = data['SKINCANCER'].apply(lambda x: cleanFeature(x))
    
    # Clean OTHERCANCER Feature
    data['OTHERCANCER'] = data['OTHERCANCER'].apply(lambda x: cleanFeature(x))
    
    # Clean CHRONICBRONCHITIS Feature
    data['CHRONICBRONCHITIS'] = data['CHRONICBRONCHITIS'].apply(lambda x: cleanFeature(x))
    
    # Clean ARTHRITIS Feature
    data['ARTHRITIS'] = data['ARTHRITIS'].apply(lambda x: cleanFeature(x))
    
    # Clean DEPRESSIVEDISORDER Feature
    data['DEPRESSIVEDISORDER'] = data['DEPRESSIVEDISORDER'].apply(lambda x: cleanFeature(x))
    
    # Clean KIDNEYDISEASE Feature
    data['KIDNEYDISEASE'] = data['KIDNEYDISEASE'].apply(lambda x: cleanFeature(x))
    
    # Clean PREGNANT Feature
    data['PREGNANT'] = data['PREGNANT'].apply(lambda x: cleanFeature(x))
    
    # Clean HYSTERECTOMY Feature
    data['HYSTERECTOMY'] = data['HYSTERECTOMY'].apply(lambda x: cleanFeature(x))

    # Clean COUGH Feature
    data['COUGH'] = data['COUGH'].apply(lambda x: cleanFeature(x))
    
    # Clean COUGHMUCUS Feature
    data['COUGHMUCUS'] = data['COUGHMUCUS'].apply(lambda x: cleanFeature(x))
    
    # Clean SHORTBREATH Feature
    data['SHORTBREATH'] = data['SHORTBREATH'].apply(lambda x: cleanFeature(x))
```

#### Engineer New Columns

```python
# Create new features
for data in [df_clean]:
    
    # Create SLEEPTIME_GROUP Feature from SLEEPTIME
    data['SLEEPTIME_GROUP'] = data['SLEEPTIME'].apply(lambda x: getSleepTimeGroup(x))
    
    # Create SEX_GROUP Feature from SEX
    data['SEX_GROUP'] = data['SEX'].apply(lambda x: getSexGroup(x))
    
    # Create WEIGHT_KILOGRAM Feature from WEIGHT
    data['WEIGHT_KILOGRAM'] = data['WEIGHT'].apply(lambda x: getWeightInKilogram(x))
    
    # Create HEIGHT_METER Feature from HEIGHT
    data['HEIGHT_METER'] = data['HEIGHT'].apply(lambda x: getHeightInMeter(x))
    
    # Create BMI Feature from WEIGHT_KILOGRAM and HEIGHT_METER    
    data['BMI'] = data['WEIGHT_KILOGRAM'] / (data['HEIGHT'] * data['HEIGHT'])
    
    # Create BMI_GROUP Feature from BMI    
    data['BMI_GROUP'] = data['BMI'].apply(lambda x: getBMIGroup(x))

    # Create RACE_GROUP Feature from RACE    
    data['RACE_GROUP'] = data['RACE'].apply(lambda x: getRaceGroup(x))
    
    # Create AGE_GROUP Feature from AGE    
    data['AGE_GROUP'] = data['AGE'].apply(lambda x: getAgeGroup(x))
```

#### Drop Columns

```python
manyMissingVals = ['ASTHMANOW', 'PREGNANT', 'HYSTERECTOMY', 'COUGH', 'COUGHMUCUS', 'SHORTBREATH', 'CANCERAGE',
                  'CANCERTYPE', 'CANCERTREATMENT']

redundantInfo = ['SLEEPTIME', 'SEX', 'WEIGHT', 'HEIGHT', 'BMI', 'RACE', 'AGE']
```

```python
for data in [df_clean]:
    
    # Drop Cols due to many missing values
    data.drop(manyMissingVals, axis = 1, inplace=True)
    
    # Drop Cols due to redundant information
    data.drop(redundantInfo, axis = 1, inplace=True)
```

#### Drop Rows 

```python
for data in [df_clean]:
    
    # Drop Rows with Nan Values in DIABETES
    data.dropna(subset=['DIABETES'], inplace=True)
    
    # Drop All Rows with Nan Values in Features
    data.dropna(inplace=True)
```
