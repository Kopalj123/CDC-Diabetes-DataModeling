# DATA VISUALIZATION

#### Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

%matplotlib inline
```

#### Read & Store Data

```python
df_clean = pd.read_csv('DiabetesCleaned_2018.csv', index_col=0) # Read csv file
```

#### Copy Data

```python
df_viz = df_clean.copy()
```

#### Display Top 5 Rows

```python
df_viz.head(10)
```

#### Create Data Visualizations

```python
fig, ((ax1, ax2, ax3), (ax4, ax5, ax6)) = plt.subplots(2,3, figsize=(30,12))

fig.suptitle('Count Bar Plots', fontsize=35)

sns.axes_style("whitegrid")

sns.countplot(x = 'SEX_GROUP', data = df_viz, hue = 'DIABETES', ax = ax1, palette = 'Paired', 
              order=['Male', 'Female'])
ax1.set_title('Patients with Diabetes by Sex', fontsize = 23)
ax1.set_yticks([x for x in range(0,300000,50000)]) 
ax1.set_yticklabels([x for x in range(0,300000,50000)], fontsize = 15)
ax1.set_ylabel('Count', fontsize = 20)
ax1.set_xticklabels(['Male', 'Female'], fontsize = 15)
ax1.set_xlabel('Sex', fontsize = 20)
ax1.legend(['Not Diabetic (0)', 'Diabetic (1)'], loc = 'upper right', fontsize = 12)

for p in ax1.patches:
    height = p.get_height()
    ax1.text(p.get_x() + p.get_width()/2.0, height + 1000,
        height , ha = "center", fontsize = 15)
    
######################################################################################################################
    
sns.countplot(x = 'SLEEPTIME_GROUP', data = df_viz, hue = 'DIABETES', ax = ax2, palette = 'Paired', 
              order=['LessSleep', 'NormalSleep', 'OverSleep'])
ax2.set_title('Patients with Diabetes by Sleeptime', fontsize = 25)

ax2.set_yticks([x for x in range(0,300000,50000)]) 
ax2.set_yticklabels([x for x in range(0,300000,50000)], fontsize = 15)
ax2.set_xlabel('Sleep Amount', fontsize = 20)
ax2.set_xticklabels(['LessSleep', 'NormalSleep', 'OverSleep'], fontsize = 15)
ax2.set_ylabel('Count', fontsize = 20)
ax2.legend(['Not Diabetic (0)', 'Diabetic (1)'], loc = 'upper right', fontsize = 12)

for p in ax2.patches:
    height = p.get_height()
    ax2.text(p.get_x() + p.get_width()/2.0, height + 1000,
        height , ha = "center", fontsize = 15)
    
######################################################################################################################
    
sns.countplot(x = 'BMI_GROUP', data = df_viz, hue = 'DIABETES', ax = ax3, palette = 'Paired',
             order=['UnderWeight', 'NormalWeight', 'OverWeight', 'Obese'])
ax3.set_title('Patients with Diabetes by BMI', fontsize = 25)

ax3.set_yticks([x for x in range(0,300000,50000)]) 
ax3.set_yticklabels([x for x in range(0,300000,50000)], fontsize = 15)
ax3.set_ylabel('Count', fontsize = 20)
ax3.set_xticklabels(['UnderWeight', 'NormalWeight', 'OverWeight', 'Obese'], fontsize = 15)
ax3.set_xlabel('BMI', fontsize = 20)
ax3.legend(['Not Diabetic (0)', 'Diabetic (1)'], loc = 'upper right', fontsize = 12)

for p in ax3.patches:
    height = p.get_height()
    ax3.text(p.get_x() + p.get_width()/2.0, height + 1000,
        height , ha = "center", fontsize = 15)
    
######################################################################################################################
    
sns.countplot(x = 'RACE_GROUP', data = df_viz, hue = 'DIABETES', ax = ax4, palette = 'Paired',
             order=['White', 'Black', 'Asian', 'AmericanIndian', 'Hispanic', 'Other'])
ax4.set_title('Patients with Diabetes by Race', fontsize = 25)

ax4.set_yticks([x for x in range(0,300000,50000)]) 
ax4.set_yticklabels([x for x in range(0,300000,50000)], fontsize = 15)
ax4.set_ylabel('Count', fontsize = 20)
ax4.set_xticklabels(['White', 'Black', 'Asian', 'AmericanIndian', 'Hispanic', 'Other'], fontsize = 15)
ax4.set_xlabel('Race', fontsize = 20)
ax4.legend(['Not Diabetic (0)', 'Diabetic (1)'], loc = 'upper right', fontsize = 12)

for p in ax4.patches:
    height = p.get_height()
    ax4.text(p.get_x() + p.get_width()/2.0, height + 1000,
        height , ha = "center", fontsize = 15)
    
######################################################################################################################

sns.countplot(x = 'AGE_GROUP', data = df_viz, hue = 'DIABETES', ax = ax5, palette = 'Paired',
             order=['YoungAdult', 'Adult', 'OldAdult', 'Elderly'])
ax5.set_title('Patients with Diabetes by Age', fontsize = 25)

ax5.set_yticks([x for x in range(0,300000,50000)]) 
ax5.set_yticklabels([x for x in range(0,300000,50000)], fontsize = 15)
ax5.set_ylabel('Count', fontsize = 20)
ax5.set_xticklabels(['YoungAdult', 'Adult', 'OldAdult', 'Elderly'], fontsize = 15)
ax5.set_xlabel('Age', fontsize = 20)
ax5.legend(['Not Diabetic (0)', 'Diabetic (1)'], loc = 'upper right', fontsize = 12)

for p in ax5.patches:
    height = p.get_height()
    ax5.text(p.get_x() + p.get_width()/2.0, height + 1000,
        height , ha = "center", fontsize = 15)
    
# Delete the last empty plot
fig.delaxes(ax = ax6) 

fig.tight_layout()

fig.savefig('CountBarPlots.png')

plt.show()
```

```python
# Calculate correlations
corr           = df_viz.corr()
matrix         = np.triu(df_viz.corr())

# Heatmap
fig, ax = plt.subplots(figsize=(12,12))
sns.set(font_scale=0.9, style="whitegrid")
# sns.axes_style("whitegrid")
sns.heatmap(corr, annot = True, ax=ax, fmt='.1g', vmin=-1, vmax=1, center= 0, 
            cbar_kws= {'fraction':0.25, 'orientation': 'vertical', "shrink": .65, 'pad': 0.05},
            annot_kws={"size": 10}, 
            square=True, mask=matrix, cmap='Blues')
ax.set_title('Heatmap Plot', fontsize = 15)
fig.savefig('HeatmapPlot.png')
```
